Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing Dataset

```python
dataset= pd.read_csv('C:\\Users\\Mayank\\Downloads\\S01 - Machine-Learning-A-Z-Codes-Datasets\\Machine-Learning-A-Z-Codes-Datasets\\Part 3 - Classification\Section 14 - Logistic Regression\\Python\\Social_Network_Ads.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset.head()
```

```python
dataset.isnull().sum()
```

Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

Training XGBoost on the Training set

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

xgb = XGBClassifier()

# 2. Define the parameter grid
parameters = {
    'n_estimators': [50, 100, 150],  # Number of trees
    'max_depth': [3, 5, 7],    # maximum depth of each individual decision tree. A higher value allows the model to learn more complex patterns , but inc. risk of overfitting    
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 1.0],  #  controls the fraction of the training data used to build each tree(random sampling),(subsample=1.0) ensures each tree is trained on the full dataset
    'colsample_bytree': [0.7, 0.8, 1.0] # controls the fraction of features (columns) that are randomly sampled for building each individual tree 
}

grid_search=GridSearchCV(estimator=xgb, param_grid=parameters, cv=5,scoring='accuracy')

# Train model with hyperparameter tunning
grid_search.fit(x_train,y_train)

```

```python
y_pred=grid_search.predict(x_test)

comparesion=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
print(comparesion)
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
